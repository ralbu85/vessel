```python
import pandas as pd
import matplotlib as mpl
mpl.rcParams['figure.dpi']= 300

raw=pd.read_csv('data/V7XJ9_10min.csv')
raw=raw.reset_index(level=0) # 시간을 기록하기 위해 인덱싱
raw=raw.dropna() # 결측치 모두 제거
import matplotlib.pyplot as plt
plt.figure(figsize=(150,5))
f, axes = plt.subplots(2, 1, figsize=(20, 10))
batch_size=int(len(raw)/2)
axcount=0
result=np.array(result)
    

for i in range(0,len(raw)-batch_size,batch_size):
    axes[axcount].grid(False)
    axes[axcount].plot(range(i,i+batch_size),raw.iloc[i:i+batch_size]['SPEED_VG'],c='black',linewidth=0.5)
    
    tmp=result[(result>=i)&(result<i+batch_size)]    
    for i in range(0,len(tmp)-1):
        if i%2==0:
            axes[axcount].axvspan(tmp[i], tmp[i+1], alpha=0.1, color='red')
        else:
            axes[axcount].axvspan(tmp[i], tmp[i+1], alpha=0.1, color='blue')        
    axcount += 1    
axes[0].axvspan(0, result[0], alpha=0.1, color='blue')
axes[0].axvspan(18000, batch_size, alpha=0.1, color='blue')
axes[1].axvspan(batch_size+1, 18485, alpha=0.1, color='blue')
axes[1].axvspan(tmp[-1], len(raw), alpha=0.1, color='red')
```
